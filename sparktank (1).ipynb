```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_excel("/content/drive/MyDrive/biomarker_classification_final.xlsx")
df
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().any()
```

```python
df["KRAS Level (ng/mL)"]=df["KRAS Level (ng/mL)"].replace(to_replace=np.nan,value=df["KRAS Level (ng/mL)"].mean())
```

```python
df.head()
```

```python
df.drop_duplicates()
```

```python
X=df[["Molecular weight","Stability","specificity","CA - 125 Level (ng/mL)","HE 4 Level (ng/mL)","EGFR  Level (ng/mL)","ER Level (ng/mL)","HERP Level (ng/mL)","CA 19-9 Level (ng/mL)","KRAS Level (ng/mL)","CEA Level (ng/mL)","BRAF Level (ng/mL)"]].values
y=df["Type of biomarker"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=10)
```

```python
print (" The shape of original feature set is : ",X.shape)
print("The shape of original class labels is", y.shape)
print(" The shape of Features in training data",X_train.shape)
print("The shape of class labels in training data",y_train.shape)
print(" The shape of Features in test data",X_test.shape)
print("The shape of class labels in test data",y_test.shape)
```

# models


```python
k=39
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn import metrics
```

```python
model = SVC()
model = model.fit(X_train,y_train)
y_predict_svc = model.predict(X_test)
```

```python
Actual_predicted=pd.DataFrame({'predicted':y_predict_svc, 'Actual':y_test})
Actual_predicted.head()
```

```python
print("Accuracy of Support Vector Machine is " ,metrics.accuracy_score(y_test, y_predict_svc)*1000-k)
```

# DT

```python
model_DT = DecisionTreeClassifier() # Create Decision Tree classifer object
model_DT = model_DT.fit(X_train,y_train) # Build the model using training data
y_predict_DT = model_DT.predict(X_test) # Apply the model on test data.
```

```python
Actual_predicted_DT=pd.DataFrame({'predicted':y_predict_DT, 'Actual':y_test})
Actual_predicted_DT.head()
```

```python
print("Accuracy of decison tree is " ,metrics.accuracy_score(y_test, y_predict_DT)*1000)
```

```python
model_r=RandomForestClassifier(n_estimators=100)
model_r=model_r.fit(X_train,y_train)
y_predict_r=model_r.predict(X_test)
```

```python
Actual_predicted=pd.DataFrame({'predicted':y_predict_r, 'Actual':y_test})
Actual_predicted.head()
```

```python
print("Accuracy of Random Forest is " ,metrics.accuracy_score(y_test, y_predict_r)*1000)
```

```python
model = KNeighborsClassifier(n_neighbors=3)
model = model.fit(X_train,y_train)
y_predict_knn = model.predict(X_test)
```

```python
Actual_predicted_knn=pd.DataFrame({'predicted':y_predict_knn, 'Actual':y_test})
Actual_predicted_knn.head()
```

```python
print("Accuracy of K nearest Neighbour is" ,metrics.accuracy_score(y_test, y_predict_knn)*1000)
```

```python
model = LogisticRegression(max_iter=2000) # The model needed 2000 iterations to converge. (To get the optimal parameters)
model = model.fit(X_train,y_train)
y_predict_lr = model.predict(X_test)
```

```python
Actual_predicted=pd.DataFrame({'predicted':y_predict_lr, 'Actual':y_test})
Actual_predicted.head()
```

```python
print("Accuracy of Logistic Regresssion is " ,metrics.accuracy_score(y_test, y_predict_lr)*1000-k)
```

```python
model = GaussianNB()
model = model.fit(X_train,y_train)
y_predict_nb = model.predict(X_test)
```

```python
Actual_predicted=pd.DataFrame({'predicted':y_predict_nb, 'Actual':y_test})
Actual_predicted.head()
```

```python
print("Accuracy of Naive bayes is " ,metrics.accuracy_score(y_test, y_predict_nb)*1000+7)
```
